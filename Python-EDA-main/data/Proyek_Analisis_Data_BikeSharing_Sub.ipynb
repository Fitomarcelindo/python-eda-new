### Cleaning Data

* cleaning dataset day

```python
# Step 1: Normalize column names
df_day.columns = df_day.columns.str.strip().str.lower()
```

```python
# Step 2: Convert date column to datetime
if 'dteday' in df_day.columns:
    df_day['dteday'] = pd.to_datetime(df_day['dteday'], errors='coerce')
```

```python
# Step 3: Check and handle missing values
missing_values = df_day.isnull().sum()
```

```python
# Filling missing values if any numerical columns have missing data
day_data_cleaned = df_day.fillna(df_day.median(numeric_only=True))
```

```python
# Step 4: Check and remove duplicate rows
duplicates = day_data_cleaned.duplicated().sum()
day_data_cleaned = day_data_cleaned.drop_duplicates()
```

```python
# Step 5: Normalize numeric columns if applicable (example for temperature)
if 'temp' in day_data_cleaned.columns:
    day_data_cleaned['temp'] = day_data_cleaned['temp'] * 41  # Scale to Celsius

if 'atemp' in day_data_cleaned.columns:
    day_data_cleaned['atemp'] = day_data_cleaned['atemp'] * 50  # Adjust scaling
```

```python
# Results summary
cleaning_results = {
    "missing_values_before": missing_values.to_dict(),
    "missing_values_after": day_data_cleaned.isnull().sum().to_dict(),
    "duplicates_removed": duplicates,
    "shape_after_cleaning": day_data_cleaned.shape
}

cleaning_results
```

## 3. Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pola penggunaan sepeda berdasarkan kondisi cuaca (weather) pada berbagai musim (season)?

```python
season_weather_data = df_day.groupby(['season', 'weathersit'])['cnt'].mean().reset_index()

# Membuat visualisasi dengan seaborn
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(
    data=season_weather_data,
    x='season',
    y='cnt',
    hue='weathersit',
    palette='viridis',
    ax=ax
)

ax.set_title('Average Bike Rentals by Weather Condition and Season')
ax.set_xlabel('Season')
ax.set_ylabel('Average Bike Rentals')

# Mengatur posisi legenda
ax.legend(title='Weather Condition', bbox_to_anchor=(1.02, 1), loc='upper left')

plt.tight_layout()  # Memastikan tampilan lebih rapi
plt.show()
```

```python
# Insights extraction
season_weather_insights = season_weather_data.sort_values(by='cnt', ascending=False).to_dict('records')
season_weather_insights
```

### Pertanyaan 2: Pada jam berapa saja penggunaan sepeda mencapai puncaknya, dan bagaimana distribusinya selama hari kerja dan akhir pekan?

```python
# Add a column to classify weekend vs weekday in the hour data
df_hour['is_weekend'] = df_hour['weekday'].apply(lambda x: 1 if x in [0, 6] else 0)
```

```python
# Calculate average bike rentals per hour during weekdays and weekends
hourly_usage = df_hour.groupby(['hr', 'is_weekend'])['cnt'].mean().reset_index()
```

```python
# Pivot the data for better visualization
hourly_usage_pivot = hourly_usage.pivot(index='hr', columns='is_weekend', values='cnt')
hourly_usage_pivot.columns = ['Weekday', 'Weekend']
```

```python
# Visualization: Hourly bike rentals distribution for weekday and weekend
plt.figure(figsize=(14, 7))
sns.lineplot(data=hourly_usage_pivot, marker="o")
```

```python
# Extract peak usage times for weekdays and weekends
peak_weekday_hour = hourly_usage_pivot['Weekday'].idxmax()
peak_weekend_hour = hourly_usage_pivot['Weekend'].idxmax()

# Generate insights dictionary for output
hourly_insights = {
    "Peak Weekday Hour": peak_weekday_hour,
    "Peak Weekend Hour": peak_weekend_hour,
    "Hourly Usage Data": hourly_usage_pivot.reset_index().to_dict('records')
}

hourly_insights
```

**Insight:**
1. Kombinasi musim dan cuaca memengaruhi rata-rata penyewaan sepeda:
   - Musim panas dengan cuaca cerah memiliki rata-rata penyewaan tertinggi.
   - Cuaca buruk seperti hujan/salju menurunkan jumlah penyewaan sepeda.
   
2. Jam sibuk penyewaan sepeda:
   - Hari kerja: Puncak penyewaan terjadi pada pukul {peak_weekday_hour}.
   - Akhir pekan: Puncak penyewaan terjadi pada pukul {peak_weekend_hour}.
   
3. Pola penggunaan berdasarkan waktu:
   - Hari kerja: Penyewaan terkait perjalanan kerja (pagi dan sore).
   - Akhir pekan: Penyewaan lebih merata dari siang hingga sore hari.
   
4. Data penggunaan per jam menunjukkan pola yang mencerminkan aktivitas pengguna:
   - Hari kerja: Penyewaan dominan pada jam komuter.
   - Akhir pekan: Penyewaan lebih banyak untuk rekreasi.
